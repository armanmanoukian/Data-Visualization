```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
global_air_pollution_dataset = pd.read_csv("global_air_pollution_dataset.csv")
global_air_pollution_dataset.head()
```

```python
global_air_pollution_dataset.describe()
```

```python
global_air_pollution_dataset['PM2.5_AQI_Value'].plot.hist(bins=30, edgecolor='black', alpha=0.5, label='Histogram', density=True, color='#8B8CF8')
ax = global_air_pollution_dataset['PM2.5_AQI_Value'].plot.kde(label='Density Plot (KDE)', color='darkred', linewidth=2)
x = ax.get_lines()[0].get_xdata()
y = ax.get_lines()[0].get_ydata()
plt.fill_between(x, y, alpha=0.3, color='darkred')
plt.xlabel('PM2.5 AQI Value')
plt.ylabel('Density')
plt.title('PM2.5 AQI Distribution with Density Overlay')
plt.legend()
plt.subplots_adjust(bottom=0.15)
plt.figtext(0.01, 0.02, 'This plot represents the distribution of PM2.5 AQI values with a density overlay.', 
            wrap=False, horizontalalignment='left', fontsize=6)

plt.show()
```

```python
global_air_pollution_dataset['PM2.5_AQI_Value'].plot.hist(bins=30, edgecolor='black', alpha=0.5, label='Histogram', density=True, color='#8B8CF8')
ax = global_air_pollution_dataset['PM2.5_AQI_Value'].plot.kde(label='Density Plot (KDE)', color='darkred', linewidth=2)
x = ax.get_lines()[0].get_xdata()
y = ax.get_lines()[0].get_ydata()
plt.fill_between(x, y, alpha=0.3, color='darkred')
plt.xlabel('PM2.5 AQI Value')
plt.ylabel('Density')
plt.title('PM2.5 AQI Distribution with Density Overlay')
plt.xlim(0, 550)
plt.xticks(np.arange(0, 600, 100))
plt.legend()
plt.subplots_adjust(bottom=0.15)
plt.figtext(0.01, 0.02, 'This plot represents the distribution of PM2.5 AQI values with a density overlay.',
            wrap=False, horizontalalignment='left', fontsize=6)

plt.show()
```

Create a Boxplot of Lung Cancer Deaths Distribution. (Python & R)

```python
plt.title('Boxplot of Lung Cancer Deaths Distribution')
sns.boxplot(y='Annual_Lung_Cancer_Deaths', data=lung_cancer_prediction_dataset, color = 'grey')
plt.title('Boxplot of Lung Cancer Deaths Distribution', color='black', fontsize=12, fontweight='bold')
plt.ylabel('Annual Lung Cancer Deaths', color='brown', fontsize=12, fontweight='bold')
plt.show();

#horizontal boxplot
plt.title('Boxplot of Lung Cancer Deaths Distribution')
sns.boxplot(x='Annual_Lung_Cancer_Deaths', data=lung_cancer_prediction_dataset, color='red')
plt.title('Boxplot of Lung Cancer Deaths Distribution', color='black', fontsize=12, fontweight='bold')
plt.xlabel('Annual Lung Cancer Deaths', color='brown', fontsize=12, fontweight='bold')
plt.show()




```
