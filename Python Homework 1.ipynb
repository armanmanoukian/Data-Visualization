```python
import pandas as pd
import numpy as np
import matplotlib as plt
# Load the dataset
crime_data = pd.read_csv(r"C:\Users\arman.manoukyan\OneDrive\Desktop\Data Visualization\Data Visulaization Homework 1\crime_data.csv")

# Check the first 5 rows
crime_data.head()
```

Identify columns with missing values and their respective counts. Drop columns where more than 50% of the data is missing (store this version as a new dataset).

```python
import pandas as pd
import os

# Identify columns with missing values and their counts
missing_values = crime_data.isnull().sum()

# Print columns with missing values
print(missing_values[missing_values > 0])

# Calculate the threshold for dropping columns (50% of the data)
threshold = len(crime_data) * 0.5

# Drop columns where more than 50% of the data is missing
crime_data_cleaned = crime_data.dropna(thresh=threshold, axis=1)

# Define a save path (e.g., in the Downloads folder)
save_path = os.path.join(os.path.expanduser("~"), "Downloads", "crime_cleanedpython.csv")

# Save the cleaned dataset
crime_data_cleaned.to_csv(save_path, index=False)

print(f"Cleaned dataset saved at: {save_path}")
```

Task 3: Convert the DATE OCC column to a datetime format. Extract the year, month, and day into separate columns. Create a new column for the hour using the TIME OCC column.

```python
import pandas as pd

# Convert 'DATE OCC' to datetime format safely
crime_data_cleaned.loc[:, 'DATE OCC'] = pd.to_datetime(
    crime_data_cleaned['DATE OCC'], format="%m/%d/%Y %I:%M:%S %p"
)

# Extract year, month, and day
crime_data_cleaned.loc[:, 'Year'] = crime_data_cleaned['DATE OCC'].dt.year
crime_data_cleaned.loc[:, 'Month'] = crime_data_cleaned['DATE OCC'].dt.month
crime_data_cleaned.loc[:, 'Day'] = crime_data_cleaned['DATE OCC'].dt.day

# --- Convert 'TIME OCC' column properly ---
# 1. Convert TIME OCC to string and pad with zeros so that each value is 4 characters long
crime_data_cleaned['TIME OCC'] = crime_data_cleaned['TIME OCC'].astype(str).str.zfill(4)

# 2. Convert the padded strings to datetime objects (using only time information)
#    This creates a datetime with a default date (e.g., 1900-01-01) so we extract just the time part.
crime_data_cleaned['TIME OCC'] = pd.to_datetime(
    crime_data_cleaned['TIME OCC'], format='%H%M'
).dt.time

# Optionally, if you still want to extract the hour from TIME OCC, you can do:
crime_data_cleaned.loc[:, 'Hour'] = crime_data_cleaned['TIME OCC'].apply(lambda t: t.hour)

# Display the first 5 rows
crime_data_cleaned.head()
```

Task 4: Filter the dataset for crimes that occurred in 2023. Further filter crimes with the description BURGLARY in the Crm Cd Desc column.

```python

# Filter for crimes in 2023
crime_data_2023 = crime_data_cleaned[crime_data_cleaned['Year'] == 2023]

# Further filter for "BURGLARY" crimes in 'Crm Cd Desc'
burglary_crimes = crime_data_2023[crime_data_2023['Crm Cd Desc'].str.contains('BURGLARY', case=False, na=False)]

# Display the first 5 rows of the filtered dataset
burglary_crimes.head()
```

Task 5: Group the data by AREA NAME and calculate the total number of crimes and the average victim age. Sort the results by total crimes in descending order.

```python
# Convert 'Vict Age' to numeric to avoid errors in aggregation
if 'Vict Age' in crime_data_cleaned.columns:
    crime_data_cleaned.loc[:, 'Vict Age'] = pd.to_numeric(crime_data_cleaned['Vict Age'], errors='coerce')

# Group by AREA NAME and calculate total number of crimes and average victim age
area_crime_stats = crime_data_cleaned.groupby('AREA NAME').agg(
    Total_Crimes=('DR_NO', 'count'),
    Avg_Victim_Age=('Vict Age', 'mean')  # Automatically ignores NaN values
).reset_index()

# Sort the results by total crimes in descending order
area_crime_stats_sorted = area_crime_stats.sort_values(by='Total_Crimes', ascending=False)

# Display the sorted results
area_crime_stats_sorted


```

Part 4: Advanced Analysis (Python & R) 20 points per problem: 10 for Python and 10 for R
1. Create a new column, Severity Score, based on the following rules:
• Assign 5 points if a weapon was used (Weapon Used Cd is not null).
• Assign 3 points for crimes under BURGLARY.
• Assign 1 point for all other crimes.
• Group by AREA NAME and find the total severity score for each area.

```python
# Make a copy of the cleaned dataset to avoid SettingWithCopyWarning
crime_data_cleaned = crime_data_cleaned.copy()

# Step 1: Create the 'Severity Score' column based on the given rules (excluding weapon usage)
crime_data_cleaned['Severity Score'] = 1  # Default: 1 point for all crimes

# Assign 3 points for BURGLARY crimes
crime_data_cleaned.loc[crime_data_cleaned['Crm Cd Desc'].str.contains('BURGLARY', case=False, na=False), 'Severity Score'] = 3

# Step 2: Group by AREA NAME and calculate the total severity score for each area
severity_by_area = crime_data_cleaned.groupby('AREA NAME').agg(
    Total_Severity_Score=('Severity Score', 'sum')
).reset_index()

# Sort by severity score in descending order
severity_by_area_sorted = severity_by_area.sort_values(by='Total_Severity_Score', ascending=False)

# Display the final results
severity_by_area_sorted
```
